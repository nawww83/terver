### 1. Случайная непрерывная величина A имеет равномерное распределение на промежутке (200, 800]. Найдите ее среднее значение и дисперсию.

Среднее значение 

$m = \dfrac{a + b}{2} = \dfrac{200 + 800}{2} = 500$

Дисперсия

$D = \dfrac{{(b-a)}^{2}}{12} = \dfrac{{(800-200)}^{2}}{12} = 30000$

## 2. О случайной непрерывной равномерно распределенной величине B известно, что ее дисперсия равна 0.2. Можно ли найти правую границу величины B и ее среднее значение зная, что левая граница равна 0.5? Если да, найдите ее.

Да, можно. Правая граница

$b = \sqrt{12 D} + a = \sqrt{12 \cdot 0.2} + 0.5 = \dfrac{2 \sqrt{3}}{5} + \dfrac{1}{2} \approx 2.05$

Среднее значение

$m = \dfrac{a + b}{2} = \dfrac{\dfrac{1}{2} + \left(\dfrac{2 \sqrt{3}}{5} + \dfrac{1}{2}\right)}{2} = \dfrac{1}{2} + \dfrac{ \sqrt{3}}{5} \approx 0.846$

## 3. Непрерывная случайная величина X распределена нормально и задана плотностью распределения f(x) = (1 / (4 * sqrt(2*pi))) * (exp(-(x+2)**2) / 32). Найдите: а). M(X) б). D(X) в). std(X) (среднее квадратичное отклонение)

Исходя из формулы для плотности вероятностей нормального распределения сразу получаем ответ:

$m = -2$

$D = \sigma^2= 4$

$\sigma = 2$

### 4.  Рост взрослого населения города X имеет нормальное распределение. Причем, средний рост равен 174 см, а среднее квадратичное отклонение равно 8 см. Какова вероятность того, что случайным образом выбранный взрослый человек имеет рост:

а). больше 182 см 

б). больше 190 см 

в). от 166 см до 190 см 

г). от 166 см до 182 см 

д). от 158 см до 190 см 

е). не выше 150 см или не ниже 190 см 

ё). не выше 150 см или не ниже 198 см 

ж). ниже 166 см.

Решение основано на интегральной функции распределения 

$\Phi(x) = \dfrac{1}{2} \left[ 1 + \text{erf} \left(\dfrac{x - m}{\sigma \sqrt{2}} \right) \right]$ 

гауссовской случайной величины. Здесь $m = 174$ см, а $\sigma = 8$ см.

а) $P = 1 - \Phi(182) \approx 0.159$

б) $P = 1 - \Phi(190) \approx 0.023$

в) $P = \Phi(190) - \Phi(166) \approx 0.819$

г) $P = \Phi(182) - \Phi(166) \approx 0.683$

д) $P = \Phi(190) - \Phi(158) \approx 0.954$

е) $P = \Phi(150) + 1 - \Phi(190) \approx 0.024$

ё) $P = \Phi(150) + 1 - \Phi(198) \approx 2.7 \cdot {10}^{-3}$

ж) $P = \Phi(166) \approx 0.159$

### 5. На сколько сигм (средних квадратичных отклонений) отклоняется рост человека, равный 190 см, от математического ожидания роста в популяции, в которой M(X) = 178 см и D(X) = 25 кв.см?

Здесь $\sigma = \sqrt{D} = \sqrt{25} = 5$ см. Разность между 190 и 178 равна 12, значит рост отклонился на $\frac{12}{5} = 2.4$ "сигм".
