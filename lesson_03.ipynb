### 1. Даны значения зарплат из выборки выпускников: 100, 80, 75, 77, 89, 33, 45, 25, 65, 17, 30, 24, 57, 55, 70, 75, 65, 84, 90, 150. Посчитать (желательно без использования статистических методов наподобие std, var, mean) среднее арифметическое, среднее квадратичное отклонение, смещенную и несмещенную оценки дисперсий для данной выборки.

Среднее арифметическое (число элементов - 20): 

$m = \dfrac{100 + 80 + 75 + 77 + 89 + 33 + 45 + 25 + 65 + 17 + 30 + 24 + 57 + 55 + 70 + 75 + 65 + 84 + 90 + 150}{20} = 65.3$

Среднее квадратичное отклонение (как корень из разности между средним квадратом и средним в квадрате):

$\sigma = \sqrt{ \dfrac{100^2+80^2+75^2+77^2+89^2+33^2+45^2+25^2+65^2+17^2+30^2+24^2+57^2+55^2+70^2+75^2+65^2+84^2+90^2+150^2}{20} - m^2 } \approx 30.82 $

Смещенная оценка дисперсии:

$D = \sigma^2 = 950.11$

Несмещенная оценка дисперсии:

$D_0 = D \cdot \dfrac{n}{n-1} = 950.11 \cdot \dfrac{20}{19} \approx 1000.12$

### 2. В первом ящике находится 8 мячей, из которых 5 - белые. Во втором ящике - 12 мячей, из которых 5 белых. Из первого ящика вытаскивают случайным образом два мяча, из второго - 4. Какова вероятность того, что 3 мяча белые?

Задачу решаем полным перебором: три белых мяча из шести могут сформироваться из следующих сочетаний:

- 11 (1000)
- 10 (1100)
- 01 (1100)
- 00 (1110)

Здесь единица в первых двух битах обозначает белый мяч из первого ящика, а в скобках обозначен вектор для результата вытаскивания из второго ящика, причем скобки говорят о том, что единицы могут сочетаться. Соответственно, в первом пункте 4 сочетания, во втором и третьем - 6, в последнем - 4.

Вероятность выпадения первого вектора равна 5/8 * 4/7 * (5/12 * 7/11 * 6/10 * 5/9) * 4, $p_1 = \frac{25}{198}$

Вероятность выпадения второго вектора равна 5/8 * 3/7 * (5/12 * 4/11 * 7/10 * 6/9) * 6, $p_2 = \frac{5}{44}$

Вероятность выпадения третьего вектора равна 3/8 * 5/7 * (5/12 * 4/11 * 7/10 * 6/9) * 6, $p_3 = p_2$

Вероятность выпадения четвертого вектора равна 3/8 * 2/7 * (5/12 * 4/11 * 3/10 * 7/9) * 4, $p_4 = \frac{1}{66}$

Искомая вероятность равна сумме четырех найденных вероятностей:

$p = p_1 + p_2 + p_3 + p_4 = \frac{73}{198} \approx 0.369$

### 3.  На соревновании по биатлону один из трех спортсменов стреляет и попадает в мишень. Вероятность попадания для первого спортсмена равна 0.9, для второго — 0.8, для третьего — 0.6. Найти вероятность того, что выстрел произведен: a). первым спортсменом б). вторым спортсменом в). третьим спортсменом.

Так как априори все стрелки могли выстрелить равновероятно (1/3), то вероятность того, что выстрел произведен, допустим, первым стрелком, пропорциональна вероятности его попадания, т.е. 0.9 или 9 (для простоты). Здесь фиксровано условие "стрелок попал", т.е. искомые вероятности являются условными. В общем случае работает формула Байеса; в нашем случае достаточно вероятности взять пропорциональными 9, 8 и 6, соответственно, так, чтобы их сумма была равна единице.

Ответ: 

$p_1 = \dfrac{9}{23}$

$p_2 = \dfrac{8}{23}$

$p_3 = \dfrac{6}{23}$

### 4.  В университет на факультеты A и B поступило равное количество студентов, а на факультет C студентов поступило столько же, сколько на A и B вместе. Вероятность того, что студент факультета A сдаст первую сессию, равна 0.8. Для студента факультета B эта вероятность равна 0.7, а для студента факультета C - 0.9. Студент сдал первую сессию. Какова вероятность, что он учится: a). на факультете A б). на факультете B в). на факультете C?

Используем формулу Байеса.

Обозначения:

Ok = событие "Студент сдал"

P(A/Ok) = вероятность того, что студент с факультета A при условии, что он сдал.

P(Ok/A) = вероятность того, что студент сдал при условии, что он с факультета A (эта вероятность задана в условии задачи и равна 0.8).

P(A) = вероятность того, что студент с факультета A. Она равна 1/4, равно как и для факультета B; для факультета C вероятность равна 1/2 (это следует из условия задачи и использования условия нормировки: $x+x+2x = 1$).

Для любого факультета A верно равенство (по сути это формула умножения вероятностей)

P(A/Ok) P(Ok) = P(Ok/A) P(A)

поэтому

P(A/Ok) = P(Ok/A) P(A) / P(Ok) = 0.8 * 1/4 / (33/40) = 8/33

Здесь P(Ok) = P(A) * P(Ok/A) + P(B) * P(Ok/B) + P(C) * P(Ok/C) = 1/4 * 0.8 + 1/3 * 0.7 + 2/4 * 0.9 = 33/40

По аналогии

P(B/Ok) = P(Ok/B) P(B) / P(Ok) = 0.7 * 1/4 / (33/40) = 7/33

P(C/Ok) = P(Ok/C) P(C) / P(Ok) = 0.9 * 2/4 / (33/40) = 18/33

Итого, ответ:

$P(\text{A/Ok}) = \dfrac{8}{33}$

$P(\text{B/Ok}) = \dfrac{7}{33}$

$P(\text{C/Ok}) = \dfrac{18}{33}$

### 5. Устройство состоит из трех деталей. Для первой детали вероятность выйти из строя в первый месяц равна 0.1, для второй - 0.2, для третьей - 0.25. Какова вероятность того, что в первый месяц выйдут из строя: а). все детали б). только две детали в). хотя бы одна деталь г). от одной до двух деталей?

а) 0.1 * 0.2 * 0.25 = 0.005

б) 0.1 * 0.2 * 0.75 + 0.1 * 0.8 * 0.25 + 0.9 * 0.2 * 0.25 = 0.08

в) 1 - 0.9 * 0.8 * 0.75 = 0.46

г) 1 - 0.9 * 0.8 * 0.75 - 0.1 * 0.2 * 0.25 = 0.455
