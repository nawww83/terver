## 1. Вероятность того, что стрелок попадет в мишень, выстрелив один раз, равна 0.8. Стрелок выстрелил 100 раз. Найдите вероятность того, что стрелок попадет в цель ровно 85 раз.

Обозначим вероятность успеха - попадания в мишень - как $p$. Тогда вероятность 85 успехов после 100 выстрелов равна числу сочетаний из 100 по 85, умноженному на вероятность успеха в степени 85, и умноженному на вероятность промаха в степени $100-85=15$. Выражая сказанное формулой, получим $C_{100}^{85} {p}^{85} {(1-p)}^{15} \approx 0.048$

## 2. Вероятность того, что лампочка перегорит в течение первого дня эксплуатации, равна 0.0004. В жилом комплексе после ремонта в один день включили 5000 новых лампочек. Какова вероятность, что ни одна из них не перегорит в первый день? Какова вероятность, что перегорят ровно две?


Обозначим вероятность перегорания в первый день экусплуатации как $p$, а число лампочек как $n$. Тогда вероятность того, что ни одна не перегорит равна ${(1-p)}^{n} = {\left( 1 - \frac{2}{5000} \right)}^{5000} \approx {e}^{-2} \approx 0.135$ (здесь использовался второй замечательный предел)


Вероятность того, что перегорят ровно две, равна ${C}_{5000}^{2} p^2 {(1 - p)}^{5000 - 2} \approx \frac{5000^2}{2} {(4 \cdot {10}^{-4})}^{2} {e}^{-2} \approx 0.27$

## 3. Монету подбросили 144 раза. Какова вероятность, что орел выпадет ровно 70 раз?

Вероятность выпадения орла равна 1/2, поэтому вероятность выпадения орла 70 раз по итогам 144 бросков равна $C_{144}^{70} {0.5}^{144} \approx 0.063$ (использовался Mathcad, функция combin(,))

## 4. В первом ящике находится 10 мячей, из которых 7 - белые. Во втором ящике - 11 мячей, из которых 9 белых. Из каждого ящика вытаскивают случайным образом по два мяча. Какова вероятность того, что все мячи белые? Какова вероятность того, что ровно два мяча белые? Какова вероятность того, что хотя бы один мяч белый?

Вероятность того, что все мячи белые равна (7/10 * 6/9) * (9/11 * 8/10), или $\frac{84}{275} \approx 0.305$

Вероятность того, что два мяча белые равна $\frac{169}{825} \approx 0.205$. Результат получен путем перебора всех случаев: если обозначить единицей 1 выпадение белого мяча, а нулем 0 - черного, то результаты последовательного вытаскивания мячей из двух ящиков, при условии выпадения ровно двух белых, можно представить в виде множества двоичных векторов {1100, 0011, 1010, 1001, 0110, 0101}. Далее для каждого вектора вычисляются вероятности, после чего - складываются.

Вероятность того, что хотя бы один белый, вычислим от обратного. Обратное событие: "Все четыре мяча - черные". Вероятность обратного события равна (3/10 * 2/9) * (2/11 * 1/10), или $\frac{1}{825} \approx 1.2 \cdot {10}^{-3}$. Отсюда искомая вероятность равна $1-\frac{1}{825} = \frac{824}{825} \approx 0.998788$
